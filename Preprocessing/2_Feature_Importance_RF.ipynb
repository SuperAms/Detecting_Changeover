```python
# Feature Selection with Univariate Statistical Tests
import pandas as pd
from pandas import read_csv
from numpy import set_printoptions
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif


```

```python
data1 = pd.read_csv('Labeled_Data_from_2021-11-26_Corr.csv', delimiter=',', index_col=0,) #Data from 31.03.2021 with Labels.csv
data2 = pd.read_csv('Labeled_Data_from_2021-12-07_Corr.csv', delimiter=',', index_col=0,) #Data from 31.03.2021 with Labels.csv

data=pd.concat([data1, data2])

print(data)
```

```python
cleaned_df = data.copy()
cleaned_df.pop('DateTime')
cleaned_df.pop('Phase_compressed')
cleaned_df.pop('Production')
#print(cleaned_df)
X=cleaned_df.iloc[:,0:cleaned_df.shape[1]-1]
print(X)
```

```python
##split data in training and test data
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(cleaned_df, test_size=0.15)

# Form np arrays of labels and features.
train_labels = np.array(train_df.pop('Phase'))
bool_train_labels = train_labels != 0
test_labels = np.array(test_df.pop('Phase'))
train_features = np.array(train_df)
test_features = np.array(test_df)
```

```python
##Z-Transformation
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
train_features = np.clip(train_features, -3, 3)
test_features = np.clip(test_features, -3, 3)


print('Training labels shape:', train_labels.shape)
print('Test labels shape:', test_labels.shape)
print('Training features shape:', train_features.shape)
print('Test features shape:', test_features.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
clf = RandomForestClassifier(n_estimators=50,random_state=0)
 
model = clf.fit(train_features,train_labels)
```

```python
##Permutationsansatz
import eli5
from eli5.sklearn import PermutationImportance
perm = PermutationImportance(model, random_state=1).fit(test_features,test_labels)
eli5.show_weights(perm, feature_names = X.columns.tolist())
```

## Alternative Package for Feature Importance

```python
from sklearn.inspection import permutation_importance
r = permutation_importance(model, test_features, test_labels,
                           n_repeats=30,
                           random_state=0)
```

```python
for i in r.importances_mean.argsort()[::-1]:
     #if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
        print(f"    {X.columns.tolist()[i]:<8}"
               f"{r.importances_mean[i]:.3f}"
               f" +/- {r.importances_std[i]:.3f}") 
```

```python
scoring = ['r2', 'neg_mean_absolute_percentage_error', 'neg_mean_squared_error']
r_multi = permutation_importance(model, test_features, test_labels, n_repeats=30, random_state=0, scoring=scoring)

for metric in r_multi:
    print(f"{metric}")
    r = r_multi[metric]
    for i in r.importances_mean.argsort()[::-1]:
        if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
            print(f"    {X.columns.tolist()[i]:<8}"
                f"{r.importances_mean[i]:.3f}"
                f" +/- {r.importances_std[i]:.3f}")
```
